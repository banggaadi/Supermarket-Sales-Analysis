```python
# Since the city and branch are basically the same i will drop the City column

data.head()
```

```python
branch_sales =data.groupby('Branch')[['Total']].sum().sort_values('Total', ascending = False)
branch_sales = branch_sales.rename(columns = {'Total': 'Total Sales'})


plots = sns.barplot(
    x = branch_sales.index,
    y = 'Total Sales',
    data = branch_sales,
)
plots.set_title('Total Sales per Branch')
for bar in plots.patches:
   
    # Using Matplotlib's annotate function and
    # passing the coordinates where the annotation shall be done
    plots.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 5),
                   textcoords='offset points')
 
```

```python
branch_profit =data.groupby('Branch')[['gross income']].sum().sort_values('gross income', ascending = False)
branch_profit = branch_profit.rename(columns = {'gross income': 'Total Profit'})

plots = sns.barplot(
    x = branch_profit.index,
    y = 'Total Profit',
    data = branch_profit,
)
plots.set_title('Total Profit per Branch')
for bar in plots.patches:
   
    # Using Matplotlib's annotate function and
    # passing the coordinates where the annotation shall be done
    plots.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 5),
                   textcoords='offset points')
```

```python
branch_sales_profit = branch_sales.merge(
    branch_profit,
    how = 'left',
    left_on = branch_sales.index,
    right_on = branch_profit.index
)

branch_sales_profit = branch_sales_profit.rename(columns = {'key_0': 'Branch'})
branch_sales_profit
```
