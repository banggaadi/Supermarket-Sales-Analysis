```python
# Since the city and branch are basically the same i will drop the City column

data.head()
```

```python
branch_sales =data.groupby('Branch')[['Total']].sum().sort_values('Total', ascending = False)
branch_sales = branch_sales.rename(columns = {'Total': 'Total Sales'})


plots = sns.barplot(
    x = branch_sales.index,
    y = 'Total Sales',
    data = branch_sales,
)
plots.set_title('Total Sales per Branch')
for bar in plots.patches:
   
    # Using Matplotlib's annotate function and
    # passing the coordinates where the annotation shall be done
    plots.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 5),
                   textcoords='offset points')
 
```

```python
branch_profit =data.groupby('Branch')[['gross income']].sum().sort_values('gross income', ascending = False)
branch_profit = branch_profit.rename(columns = {'gross income': 'Total Profit'})

plots = sns.barplot(
    x = branch_profit.index,
    y = 'Total Profit',
    data = branch_profit,
)
plots.set_title('Total Profit per Branch')
for bar in plots.patches:
   
    # Using Matplotlib's annotate function and
    # passing the coordinates where the annotation shall be done
    plots.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 5),
                   textcoords='offset points')
```

```python
branch_sales_profit = branch_sales.merge(
    branch_profit,
    how = 'left',
    left_on = branch_sales.index,
    right_on = branch_profit.index
)

branch_sales_profit = branch_sales_profit.rename(columns = {'key_0': 'Branch'})
branch_sales_profit
```

#### 2.Sales Trends Over Time:

* What are the peak sales times during the day?
* Are there any seasonal trends in sales?
* How do sales vary by day of the week or month?

```python
data.head()
```

```python
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M').dt.time
data['Hour'] = pd.to_datetime(data['Time'], format='%H:%M:%S').dt.hour
```

```python
hourly = data.groupby(['Hour', 'Branch'])[['Total']].sum().reset_index()

sns.relplot(x='Hour', y='Total', data=hourly, kind='line', hue ='Branch')

plt.ylabel('Total Sales')
plt.title('Hourly Sales')
for branch in hourly['Branch'].unique():
    branch_data = hourly[hourly['Branch'] == branch]
    peak_row = branch_data.loc[branch_data['Total'].idxmax()]
    lowest_row = branch_data.loc[branch_data['Total'].idxmin()]
    plt.text(
        x=peak_row['Hour'], 
        y=peak_row['Total'], 
        s=f'{peak_row["Total"]:.2f}', 
        color='black', 
        ha='center', 
        va='bottom'
    )
    plt.text(
        x=lowest_row['Hour'], 
        y=lowest_row['Total'], 
        s=f'{lowest_row["Total"]:.2f}', 
        color='black', 
        ha='center', 
        va='bottom'
    )


# Display the plot
plt.show()
```

```python
seasonal = data[['Invoice ID', 'Branch', 'Total', 'Date']]
seasonal['Month'] = data['Date'].dt.month
seasonal['DayofWeek'] = data['Date'].dt.dayofweek
```

```python
tes = seasonal.groupby(['Branch', 'Month', 'DayofWeek'])[['Total']].sum().reset_index()
tes
```

```python
seasonal = data[['Invoice ID', 'Branch', 'Total', 'Date']].copy()

seasonal['Month'] = seasonal['Date'].dt.month
seasonal['DayOfWeek'] = seasonal['Date'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
seasonal['DayOfWeek'] = pd.Categorical(seasonal['DayOfWeek'], categories=day_order, ordered=True)

tes_pivot = seasonal.pivot_table(
    index='Branch',
    columns=['Month', 'DayOfWeek'],
    values='Total',
    aggfunc='sum'
)

unique_months = sorted(seasonal['Month'].unique())

for month in unique_months:
    plt.figure(figsize=(8, 4))
    
    month_data = tes_pivot.xs(month, level='Month', axis=1)
    
    month_data = month_data.reset_index()
    
    month_data = month_data.melt(id_vars='Branch', var_name='DayOfWeek', value_name='Total')
    
    sns.lineplot(x='DayOfWeek', y='Total', hue='Branch', data=month_data, marker='o')

    plt.title(f'Total Sales by Branch and Day of the Week - Month {month}')
    plt.xlabel('Day of the Week')
    plt.ylabel('Total Sales')
    plt.legend(title='Branch', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=0)

    plt.show()
```

```python
heatmap_data = seasonal.pivot_table(
    index='DayOfWeek',
    columns='Branch',
    values='Total',
    aggfunc='sum'
).reindex(day_order)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlGnBu")

plt.title('Total Sales by Branch and Day of the Week')
plt.xlabel('Branch')
plt.ylabel('Day of the Week')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='DayOfWeek', y='Total', hue='Branch', data=seasonal, order=day_order)

plt.title('Total Sales Distribution by Branch and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.legend(title='Branch', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)

plt.show()
```

```python
tester = seasonal.groupby(['Branch', 'Month', 'DayOfWeek'])[['Total']].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.boxplot(x='DayOfWeek', y='Total', hue='Branch', data=tester, order=day_order)

plt.title('Total Sales Distribution by Branch and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.legend(title='Branch', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)

plt.show()
```

#### 3.Product Sales Analysis:

* Which product line has the highest sales?
* Which product lines generate the highest profit?
* Which product lines are popular between different genders?

```python
Question_3 = data[['Product line', 'Gender', 'Unit price', 'Quantity', 'Total','cogs', 'gross income']]
Question_3
```

```python
Question_3.groupby('Product line')[['Total']].sum().sort_values('Total', ascending = False)
```

```python
Question_3.groupby('Product line')[['gross income']].sum().sort_values('gross income', ascending = False)
```

```python
stacked_df = pd.DataFrame((Question_3.groupby(['Product line', 'Gender'])[['Total']].sum()).reset_index())
stacked_pivot = stacked_df.pivot_table(
    index = 'Product line',
    columns = 'Gender',
    values = 'Total'
)
```

```python
fig ,ax = plt.subplots()

width = .25
x = np.arange(0, len(Question_3['Product line'].unique()))

ax.bar(
    x-width/2,
    stacked_pivot['Male'],
    width = width,
    label = 'Male'
)

ax.bar(
    x+width/2,
    stacked_pivot['Female'],
    width = width,
    label = 'Female'
)

ax.set_title('Male vs Female')
ax.set_ylabel('Revenue')
ax.set_xticks(x)
ax.set_xticklabels(stacked_pivot.index, rotation = 45 )

ax.legend()

plt.tight_layout() 
```

```python
pie = Question_3.groupby('Gender')[['Total']].sum()
fig, ax = plt.subplots()

ax.pie(
    x = pie['Total'],
    startangle = 90,
    labels = pie.index,
    autopct = '%0.f%%'
)

ax.set_title('Customer by Gender')
```
